```python
import pandas as pd
import sys
sys.path.append('..')
from src.rag import ProductRAG

#Load dataset
df = pd.read_csv('../data/Product_Information_Dataset.csv')
df.head()
```

```python
#Initiate RAG & Create vectorstore
rag = ProductRAG(model_name='all-MiniLM-L6-v2')
rag.create_vectorstore(df)
print('Vectorstore created successfully')
```

```python
def printResults(docs):
    for i, doc in enumerate(docs, 1):
        print(doc.page_content)

# Prueba with more specific queries
queries = [
    'What are the top 5 highly-rated guitar products?',
    'What is a good product for thin guitar strings?',
    'Is the BOYA BYM1 Microphone good for a cello?',
]

for query in queries:
    print(f'\n\n Search: "{query}"')
    print('=' * 50)
    resultados = rag.search(query, k=5)
    printResults(resultados)
```
